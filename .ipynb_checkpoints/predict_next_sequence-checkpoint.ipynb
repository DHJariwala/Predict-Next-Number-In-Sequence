```python
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
data = [10*a for a in range(1,21)]
data
```

```python
def splitSequence(seq,n_steps):
    x = []
    y = []
    for i in range(len(seq)):
        lastIndex = i + n_steps
        if lastIndex > len(seq) - 1:
            break
        seqx,seqy = seq[i:lastIndex],seq[lastIndex]
        x.append(seqx)
        y.append(seqy)
        pass
    x = np.array(x)
    y = np.array(y)
    return x,y
```

```python
n_steps = 5
X,y = splitSequence(data,n_steps=5)
```

```python
print(X)
print(y)
```

```python
for i in range(len(X)):
    print(X[i],y[i])
```

```python
# Reshape data
n_features = 1
X = X.reshape((X.shape[0],X.shape[1],n_features))

print(X[:2])
```

```python
# Sequential model
model = tf.keras.Sequential()
model.add(layers.LSTM(50,activation='relu',input_shape=(n_steps,n_features)))
model.add(layers.Dense(1))
```

```python
model.layers
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss=tf.keras.losses.MeanSquaredError(),metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=200,verbose=1)
```

```python
# Prediction#1
num = 120
arr = [a for a in range(num,num+50,10)]
test_data = np.array(arr)
test_data = test_data.reshape((1,n_steps,n_features))
test_data
```

```python
predictNextNumber = model.predict(test_data,verbose=1)
print(arr,predictNextNumber)
```

```python
# Prediction#2
num = 120
arr = [a for a in range(num,num+50,10)]
test_data = np.array(arr)
test_data = test_data.reshape((1,n_steps,n_features))
predictNextNumber = model.predict(test_data,verbose=1)
print(arr,predictNextNumber)
```
